Carga de Datasets 

```python

orders = pd.read_csv("/home/tomas/Desktop/project-sales-analysis/data/raw/archive/olist_orders_dataset.csv")
order_items = pd.read_csv("/home/tomas/Desktop/project-sales-analysis/data/raw/archive/olist_order_items_dataset.csv")
customers = pd.read_csv("/home/tomas/Desktop/project-sales-analysis/data/raw/archive/olist_customers_dataset.csv")
payments = pd.read_csv("/home/tomas/Desktop/project-sales-analysis/data/raw/archive/olist_order_payments_dataset.csv")
reviews = pd.read_csv("/home/tomas/Desktop/project-sales-analysis/data/raw/archive/olist_order_reviews_dataset.csv")
products = pd.read_csv("/home/tomas/Desktop/project-sales-analysis/data/raw/archive/olist_products_dataset.csv")
sellers = pd.read_csv("/home/tomas/Desktop/project-sales-analysis/data/raw/archive/olist_sellers_dataset.csv")
geo = pd.read_csv("/home/tomas/Desktop/project-sales-analysis/data/raw/archive/olist_geolocation_dataset.csv")
```

```python
# Unir ordenes con ítems
df = orders.merge(order_items, on="order_id", how="inner")

# Agregar clientes
df = df.merge(customers, on="customer_id", how="left")

# Agregar pagos
df = df.merge(payments, on="order_id", how="left")

# Agregar reviews
df = df.merge(reviews, on="order_id", how="left")

# Agregar productos
df = df.merge(products, on="product_id", how="left")

# Agregar vendedores
df = df.merge(sellers, on="seller_id", how="left")
```
